#### 4. Improve The Quality of Data

##### a. Drop off missing value and unnessary column:
- For Size column, since the fraction missing is almost 70%, we have to drop the whole column.
- For region & state columns, it would be better to drop off those column since we neglect the price different between states in US.
- In addition, such columns like VIN and Id do not affect the price of the used car
- Finally, let drop all rows with missing data

```python
#Drop size column
raw_df = raw_df.drop(['size', 'id', 'VIN', 'region', 'state'],axis=1);

#Drop off all row with missing values
raw_df = raw_df.dropna();
```

##### b. Eliminate the outliers:
- Adjust the price range to make more sense for used car (1500 - $300000).
- Adjust the odometer range to make more sense for used car (above 10000 miles).
- Applied method interquatile range to eliminate the outilers in numeric columns

```python
# The prince range assume to be between $1500 - $300000 for used car
raw_df = raw_df[(raw_df['price']>1500) & (raw_df['price']<300000)]
# The odometer range assume to be above 5000 miles
raw_df = raw_df[raw_df['odometer']>10000]
```

```python
numeric_cols = raw_df.select_dtypes(include=np.number).columns.tolist();
for col in numeric_cols:
    Q1 = raw_df[col].quantile(0.25);
    Q3 = raw_df[col].quantile(0.75);
    IQR = Q3 - Q1;
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    no_outlier_df = raw_df[(raw_df[col] >= lower_bound) & (raw_df[col] <= upper_bound)]
```

### Data Preparation

After our initial exploration and fine-tuning of the business understanding, it is time to construct our final dataset prior to modeling.  Here, we want to make sure to handle any integrity issues and cleaning, the engineering of new features, any transformations that we believe should happen (scaling, logarithms, normalization, etc.), and general preparation for modeling with `sklearn`. 

```python
no_outlier_df.info()
```

```python
input_df = no_outlier_df.drop(['price','model'], axis=1);
num_cols = input_df.select_dtypes(include=np.number).columns.tolist();
categorical_cols = ['manufacturer', 'condition','cylinders', 'fuel','title_status','transmission','drive','type','paint_color']
target_col = 'price';
target = no_outlier_df[target_col];
```

```python
input_df.head()
```

#### 1. Apply James Stein Encoder
- We apply James Stein Encoder to the following columns: manufacturer, cylinders, type, and paint colors because these columns have a large number of unique values. It's reasonable to replace with the avereage price because each feature has relationship to the target price

```python
# Initialize the JamesSteinEncoder with the column to encode
JS_encoder = ce.JamesSteinEncoder(cols=['manufacturer','cylinders','type','paint_color']);

# Fit and transform the data
input_JSencoded = JS_encoder.fit_transform(input_df, target);

# Show the encoded training data
input_JSencoded.head()
```

#### 2. Create Column Transformer 
- Create columns transformer to apply Ordinal and OneHot Encoder to the following categorical columns

```python
transformer = make_column_transformer(
        (OrdinalEncoder(), ['condition']),
        (OneHotEncoder(), ['fuel','title_status','transmission','drive']),
        remainder='passthrough'
)
```

#### 3. Create training and test set

```python
X_train, X_test, y_train, y_test = train_test_split(input_JSencoded, target, test_size=0.3, random_state=42);
X_train.head()
```

### Modeling

With your (almost?) final dataset in hand, it is now time to build some models.  Here, you should build a number of different regression models with the price as the target.  In building your models, you should explore different parameters and be sure to cross-validate your findings.

#### 1. Base Model with Default Linear Regression

```python
basemodel = Pipeline([
    ('transform', transformer),
    ('linreg', LinearRegression())
]);
basemodel.fit(X_train, y_train);
basemodel.score(X_test, y_test);

base_y_testPred = basemodel.predict(X_test);

base_mse = mean_squared_error(y_test, base_y_testPred);
base_score = basemodel.score(X_test,y_test);
print(f"Mean Squared Error of the base model: {base_mse}")
print(f"Score of the base model: {base_score}")
```

#### 2. Model with GridSearch to find optimal polyfeature

```python
train_mses = []
test_mses = []
test_score = []
for i in range(1,6):
    poly_transform = make_column_transformer(
        (PolynomialFeatures(degree=i, include_bias=False),['year','odometer','manufacturer','cylinders','type','paint_color']),
        (OrdinalEncoder(), ['condition']),
        (OneHotEncoder(), ['fuel','title_status','transmission','drive']),
        remainder='passthrough'
    )
    pipe = Pipeline([
        ('transform', poly_transform),
        ('linreg', LinearRegression())
    ])
    pipe.fit(X_train, y_train);
    train_preds = pipe.predict(X_train);
    test_preds = pipe.predict(X_test);
    train_mses.append(mean_squared_error(y_train, train_preds));
    test_mses.append(mean_squared_error(y_test, test_preds));
    test_score.append(pipe.score(X_test,y_test));

best_model_complexity = test_mses.index(min(test_mses)) + 1;
```

```python
print(f'The Complexity that minimized Test Error was: {test_mses.index(min(test_mses)) + 1}')
print(f'The Best MSE of Test Error was: {min(test_mses)}')
print(f'The Best Score of Test Set was: {max(test_score)}')
plt.plot(range(1,6), train_mses, '--o', label = 'training error')
plt.plot(range(1,6), test_mses, '--o', label = 'testing error')
plt.xticks(range(1, 6), range(1, 6))
plt.xlabel('Degree Complexity')
plt.ylabel('Mean Squared Error')
plt.legend();
```

#### 3. Model with GridSearch + Ridge Regresion to find optimal alpha

```python
poly_transform = make_column_transformer(
        (PolynomialFeatures(degree=2, include_bias=False),['year','odometer','manufacturer','cylinders','type','paint_color']),
        (OrdinalEncoder(), ['condition']),
        (OneHotEncoder(), ['fuel','title_status','transmission','drive']),
        remainder='passthrough')

param_dict = {'ridge__alpha': [0.0001, 0.001, 0.1, 1.0, 10.0, 100.0, 1000]}

ridgemodel = Pipeline([
    ('transformer', poly_transform),
    ('Std_scaler', StandardScaler()),
    ('ridge', Ridge())
]);

# Create the GridSearchCV object with 5-fold cross-validation
grid_search = GridSearchCV(estimator=ridgemodel, param_grid=param_dict, cv=5, scoring='neg_mean_squared_error')

# Fit the grid search to the training data
grid_search.fit(X_train, y_train)

# Get the best alpha value found by GridSearchCV
best_alpha = grid_search.best_params_['ridge__alpha']
print(f"Best alpha value: {best_alpha}")

# Get the best model (Lasso with the best alpha)
best_model = grid_search.best_estimator_

# Make predictions on the test set
y_pred = best_model.predict(X_test)

# Evaluate the model
ridge_mse = mean_squared_error(y_test, y_pred)
ridge_score = best_model.score(X_test,y_test);
print(f"Mean Squared Error of the best model: {ridge_mse}")
print(f"Score of the best model: {ridge_score}")
```

#### 5. Apply Sequential Feature Selection with Poly Features + Ridge Model to define the best number of feature

```python
train_mses = []
test_mses = []
test_score = []
poly_transform = make_column_transformer(
        (PolynomialFeatures(degree=2, include_bias=False),['year','odometer','manufacturer','cylinders','type','paint_color']),
        (OrdinalEncoder(), ['condition']),
        (OneHotEncoder(), ['fuel','title_status','transmission','drive']),
        remainder='passthrough'
    )
for i in range(1,21):
    SFS_model = Pipeline([
        ('transformer', poly_transform),
        ('Std_scaler', StandardScaler()),
        ('selector', SequentialFeatureSelector(LinearRegression(), direction='forward', n_features_to_select=i)),
        ('ridge', Ridge(alpha=0.001))
    ]);
    SFS_model.fit(X_train, y_train);
    train_preds = SFS_model.predict(X_train);
    test_preds = SFS_model.predict(X_test);
    train_mses.append(mean_squared_error(y_train, train_preds));
    test_mses.append(mean_squared_error(y_test, test_preds));
    test_score.append(SFS_model.score(X_test,y_test));

best_n_selector= test_score.index(max(test_score)) + 1;
print(f"Number of features with highest accuracy score: {best_n_selector}")
print(f"Highest accuracy score: {max(test_score)}")
```

```python
plt.plot(range(1,21), test_score, '--o', label = 'test score')
plt.xticks(range(1, 21), range(1, 21))
plt.xlabel('Number of Selected Features')
plt.ylabel('Score')
plt.legend();
```

**Conclusion:** 
- The features in the dataset has nonlinear characteristic since the accuracy score of the poly feature with degree of 2 is significantly higher than the base model with linear feature.
- The model is also tuned and improved by adjusting the alpha value in Ridge Model to increase the accuracy score of test set.
- After investigate number of feature using Sequential Features Selection, there was a significant increase of test set score up to 15 features. After that, the test set score flattened out.

### Evaluation

With some modeling accomplished, we aim to reflect on what we identify as a high-quality model and what we are able to learn from this.  We should review our business objective and explore how well we can provide meaningful insight into drivers of used car prices.  Your goal now is to distill your findings and determine whether the earlier phases need revisitation and adjustment or if you have information of value to bring back to your client.

#### 1. Train the final model

```python
final_transform = make_column_transformer(
        (PolynomialFeatures(degree=2, include_bias=False),['year','odometer','manufacturer','cylinders','type','paint_color']),
        (OrdinalEncoder(), ['condition']),
        (OneHotEncoder(), ['fuel','title_status','transmission','drive']),
        remainder='passthrough')


finalmodel = Pipeline([
    ('transformer', final_transform),
    ('Std_scaler', StandardScaler()),
    ('selector', SequentialFeatureSelector(LinearRegression(), direction='forward', n_features_to_select=15)),
    ('ridge', Ridge(alpha=0.001))
]);

finalmodel.fit(X_train, y_train);

train_preds = finalmodel.predict(X_train);
test_preds = finalmodel.predict(X_test);
final_mse = mean_squared_error(y_test, test_preds);
final_score = finalmodel.score(X_test,y_test);

print(f"Mean Squared Error of the Final model: {final_mse}")
print(f"Score of the Final model: {final_score}")
```

#### 2. Investigate the important of features

```python
#Calculate the permutation importance
results = permutation_importance(finalmodel, X_test, y_test);
results = pd.DataFrame(data=results.importances_mean, index=input_JSencoded.columns, columns=['Importance']).sort_values(by='Importance', ascending=False)
results.head(10)
```

```python
#Calculate the permutation importance
results = permutation_importance(finalmodel, X_train, y_train);
results = pd.DataFrame(data=results.importances_mean, index=input_JSencoded.columns, columns=['Importance']).sort_values(by='Importance', ascending=False)
results.head(10)
```
